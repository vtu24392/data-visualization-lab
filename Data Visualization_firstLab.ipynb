```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
x  = [2, 4, 6, 8, 10, 12, 13, 14, 16, 18]
y1 = [1, 3, 5, 3, 2, 3, 5, 2, 3, 1]
y2 = [2, 4, 6, 3, 1, 4, 6, 5, 2, 4]
y3 = [5, 3, 2, 6, 4, 1, 5, 2, 3, 2]
plt.plot(x, y1, label="line L")
plt.plot(x, y2, label="line H")
plt.plot(x, y3, label="Line S")
plt.plot()

plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("Line Graph Example")
plt.legend()
plt.show()
```

```python
# Sample data
data = {
    'Month' : ['Apr','May', 'Jun', 'Jul', 'Aug', 'Sep'],
    'Sales' : [2, 3, 4, 5, 3, 7]
}
df = pd.DataFrame(data)
df.plot(x='Month', y='Sales', title='Monthly sales in FY 25-26')
plt.xlabel("Month")
plt.ylabel("Sales (in million units)")
plt.grid(visible='True')
plt.show()
```

```python
sns.lineplot(x="Month", y="Sales", color="red", data=df).set_title("Monthly sales in FY 25-26")
sns.barplot(x="Month", y="Sales", data=df).set_title("Monthly sales in FY 25-26")


plt.show()
```
