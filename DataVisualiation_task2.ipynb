##1. Categorical Data - Bar chart, Pie chart

```python
# Import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('StudentsPerformance.csv')
```

```python
print(df['gender'].value_counts())
print(df['gender'].value_counts(normalize=True) * 100) # Percentage
```

```python
#Bar Chart
sns.countplot(x='gender', data=df)
plt.title("Bar Chart - Gender Distribution")
plt.xlabel("Gender")
plt.ylabel("Count")
plt.show()
```

```python
# Prepare data
gender_counts = df['gender'].value_counts()
# Plot pie chart
plt.pie(gender_counts, labels=gender_counts.index, autopct='%1.1f%%', startangle=90)
plt.title("Pie Chart - Gender Distribution")
plt.axis('equal') # Equal aspect ratio ensures pie is a circle
plt.show()
```

##2. Continuous Data - common visualizations


* Histogram
* Box Plot
* Line Plot
* Density Plot
* Rug Plot
* Strip Plot
* Swarm Plot



```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("StudentsPerformance.csv")
```

```python
print(df.head())
```

```python
math_scores = df['math score']
reading_scores = df['reading score']
writing_scores = df['writing score']

# Summary statistics
print("Math Score Summary:")
print(math_scores.describe())
print("\nReading Score Summary:")
print(reading_scores.describe())
print("\nWriting Score Summary:")
print(writing_scores.describe())
```

```python
# Histogram plot
sns.histplot(math_scores, bins=10, kde=False)
plt.title("Histogram of Math Scores")
plt.xlabel("Math Score")
plt.ylabel("Frequency")
plt.show()
```

```python
# Box plot
sns.boxplot(x=math_scores)
plt.title("Box Plot of Math Scores")
plt.xlabel("Math Score")
plt.show()
```

```python
# Density plot
sns.kdeplot(math_scores, fill=True)
plt.title("Density Plot of Math Scores")
plt.xlabel("Math Score")
plt.show()
```

```python
# Rug plot
sns.rugplot(math_scores, height=0.1)
plt.title("Rug Plot of Math Scores")
plt.xlabel("Math Score")
plt.show()
```
