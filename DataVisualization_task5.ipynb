### 5.a. Create TreeMap and Sunburst charts with International Financial Statistics dataset which contain data related to different product categories, their subcategories, and the revenue generated by each product in different regions.

```python
import plotly.express as px
import pandas as pd
```

```python
data = {
    'Region': ['North America', 'North America', 'Europe', 'Europe', 'Asia', 'Asia'],
    'Category': ['Electronics', 'Clothing', 'Electronics', 'Clothing', 'Electronics', 'Clothing'],
    'Subcategory': ['Smartphones', 'Shoes', 'Laptops', 'T-Shirts', 'Tablets', 'Jeans'],
    'Revenue': [5000, 2000, 8000, 1500, 7000, 1200]
}
```

```python
df = pd.DataFrame(data)
```

```python
fig_tree_map = px.treemap(df, path=['Region', 'Category', 'Subcategory'], values='Revenue', title='TreeMap Chart: Revenue by Region, Category, and Subcategory')
fig_tree_map.show()
```

```python
fig_sunburst = px.sunburst(df, path=['Region', 'Category', 'Subcategory'], values='Revenue', title='Sunburst Chart: Revenue by Region, Category, and Subcategory')
fig_sunburst.show()
```

### 5.b) Display Sunburst charts with data related to different departments in the university, their students with top 5 in their academics, sports, placement, etc.

```python
import plotly.graph_objects as go
```

```python
data = {
    'Department': ['Engineering', 'Engineering', 'Engineering', 'Engineering', 'Engineering','Science', 'Science', 'Science', 'Science', 'Science'],
    'Category': ['Academics', 'Academics', 'Sports', 'Sports', 'Placement', 'Academics', 'Academics', 'Sports', 'Sports', 'Placement'],
    'Student': ['John', 'Emma', 'Mike', 'Sarah', 'David', 'Alice', 'Bob', 'Olivia', 'Jack', 'Sophia'],
    'Score': [95, 92, 85, 80, 95, 93, 91, 87, 83, 95]
}
# Create DataFrame
df = pd.DataFrame(data)
# Filter top 5 students for each department and category
top_students = df.groupby(['Department', 'Category']).apply(lambda x: x.nlargest(5, 'Score')).reset_index(drop=True)
```

```python
fig = go.Figure(go.Sunburst(labels=top_students['Department'] + ' - ' + top_students['Category'] + ': ' + top_students['Student'], parents=[''] * len(top_students), values=top_students['Score']))
fig.update_layout(title='Top 5 Students in Academics, Sports, and Placement by Department')
fig.show()
```
